```python
import pandas as pd
```

```python
df = pd.read_csv("titanic-passengers.csv", sep=";", encoding="ISO8859-1")
df
```

```python
df["Survived"]=df["Survived"].map({"Yes": 1, "No": 0})
df
```

```python
df.isnull().sum()
```

```python
df.drop('Cabin',axis=1,inplace=True)
df
```

```python
df.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
df
```

```python
df_dummies = pd.get_dummies(df[['Sex','Embarked']],drop_first=True)
df
```

```python
df.drop(['Sex','Embarked'],axis=1,inplace=True)
df
```

```python
df1 = pd.concat([df,df_dummies],axis=1)
df1
```

```python
df1.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics


x = df1[['Pclass', 'Fare']]
y = df1['Survived']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0) 

logreg = LogisticRegression()  
logreg.fit(x_train, y_train) 
y_pred  = logreg.predict(x_test)    
print("Accuracy={:.2f}".format(logreg.score(x_test, y_test)))
```

```python
import seaborn as sns
sns.regplot(x='Fare',y='Survived',data=df1)
```

```python
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(scaled,y,test_size=0.30,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt
y_pred_proba = model.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="Titanic, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
